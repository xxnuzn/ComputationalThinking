```python
# 두 수를 더하는 프로그램

def get_input():
    """사용자로부터 두 숫자를 입력받는 함수"""
    x = float(input("첫 번째 숫자를 입력하세요: "))
    y = float(input("두 번째 숫자를 입력하세요: "))
    return x, y

def add_numbers(x, y):
    """두 숫자를 더하는 함수"""
    return x + y

def display_result(result):
    """결과를 출력하는 함수"""
    print(f"두 숫자의 합은: {result}")

# 문제를 분해하여 각 부분을 처리
x, y = get_input()
result = add_numbers(x, y)
display_result(result)
```
